```python
# UnitPrice 18'den büyük olan verilerin ProductID, ProductName, UnitPrice, UnitsInStock sütunlarını listeleyin
# Path I
(df.query('UnitPrice > 18.00')
    .loc[:, ['ProductID', 'ProductName', 'UnitPrice', 'UnitsInStock']]
    .sort_values(by='UnitPrice', ascending=False))
# Path II
# df[
#     df['UnitPrice'] > 18.00
# ][['ProductID', 'ProductName', 'UnitPrice', 'UnitsInStock']].sort_values(by='UnitPrice', ascending=False)

# Path III
# df[
#     df['UnitPrice'] > 18.00
# ].loc[:, ['ProductID', 'ProductName', 'UnitPrice', 'UnitsInStock']].sort_values(by='UnitPrice', ascending=False)
```

```python
df_fuel_consumption= pd.read_csv(filepath_or_buffer='data/FuelConsumption.csv')
df_fuel_consumption[:10]
```

```python
df_fuel_consumption.shape
```

```python
df_fuel_consumption.dtypes
```

```python
# groups.keys() MAKE isimlerini gösterir
df_fuel_consumption.groupby(['MAKE']).groups.keys()
```

```python
# groups fonksiyonu index numalarını gösterir
df_fuel_consumption.groupby(['MAKE']).groups 
```

```python
# MAKE' da JAGUAR olanlar, CYLINDERS 8 olan FUELTYPE Z olanların MAKE, VEHICLECLASS, ENGINESIZE, CYLINDERS, TRANSMISSION, FUELTYPE listeleyelim

(df_fuel_consumption.query('MAKE.str.contains("JAGUAR") and CYLINDERS == 8 and FUELTYPE == "Z"')
    .loc[:, ['MAKE', 'VEHICLECLASS', 'ENGINESIZE', 'CYLINDERS', 'TRANSMISSION', 'FUELTYPE']]
    .sort_values(by='CYLINDERS', ascending = False)
 )
```

```python
# Markalara göre FUELCONSUMPTION_HWY ortalamalarını alalım

df_fuel_consumption.groupby('MAKE')[['FUELCONSUMPTION_HWY']].mean()
```

```python
# Markalara göre  FUELCONSUMPTION_HWY, FUELCONSUMPTION_COMB, ortalama ve orta değerlerini listeleyelim

df_fuel_consumption.groupby('MAKE')[['FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB']].agg(['mean', 'median'])
```

```python
# Yakıt tüketimi az olan ilk 5 aracı Marka, Model ve Yakıt ortalamasını alıp sıralayım
(df_fuel_consumption
 .groupby(['MAKE', 'MODEL'])[['FUELCONSUMPTION_COMB']]
 .agg(['mean', 'median'])
 .sort_values(('FUELCONSUMPTION_COMB', 'mean'))
 .head(5)
)
```

```python

# Bütün markaları ve tüm değerlerini listeleyelim

df_fuel_consumption.groupby('MAKE').first()
```

```python

# VOLVO markasının tüm değerlerini getirelim.
df_fuel_consumption[df_fuel_consumption['MAKE'] == 'VOLVO']
```

```python
# Araçların FUELTYPE'nin FUELCONSUMPTION_CITY, FUELCONSUMPTION_HWY, FUELCONSUMPTION_COMB, FUELCONSUMPTION_COMB_MPG özelliklerini mean ve median sıralayalım
df_fuel_consumption.groupby('FUELTYPE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']] \
                   .agg(['mean', 'median'])
```

```python
# Araçarın Apply fonksiyonu ile  3.5 ile 6.5 ENGINESIZE arasındaki mean, median, min, max değerlerini FUELCONSUMPTION_CITY, FUELCONSUMPTION_HWY, FUELCONSUMPTION_COMB, FUELCONSUMPTION_COMB_MPG, CO2EMISSIONS listeleyelim
# agg groupby nesnesi üstünde toplu işlem yapmayı sağlar ve return ile birlikte bunları geri döndürür
# avg ortalama almayı sağlar
# apply veriyi sırayla parçalar ve bu parçalara bir fonksiyon uygular
def min_max_mean_median(avg): 
    return avg.agg(['mean','median', 'min', 'max']) # toplu işlem yapmayı sağlar ve içinde döndürür

(df_fuel_consumption.query('ENGINESIZE >= 3.5 and ENGINESIZE <= 6.5')
    .groupby('ENGINESIZE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
    .apply(min_max_mean_median)
)
```

```python
# Araçarın Apply fonksiyonu ile  3.5 ile 6.5 ENGINESIZE arasındaki mean, median, min, max değerlerini FUELCONSUMPTION_CITY, FUELCONSUMPTION_HWY, FUELCONSUMPTION_COMB, FUELCONSUMPTION_COMB_MPG, CO2EMISSIONS listeleyelim
# apply() yerine agg() kullandık agg() daha sade bir çözümdür.
# apply() kullanmak, her gruba özel bir işlem yapmak isteseydin (örneğin: belirli koşula göre sadece min-max döndürmek), apply() devreye girer
# agg() bu iş için en uygun ve okunabilir yöntem.
#  apply() ise özel durumlar ve esnek kontroller için.
(df_fuel_consumption.query('ENGINESIZE >= 3.5 and ENGINESIZE <= 6.5')
    .groupby('ENGINESIZE')[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]
    .agg(['mean', 'median', 'min', 'max'])
)
```

```python
# AUDI modellerini FUELCONSUMPTION_HWY azdan çoka sıralayım.
df_fuel_consumption[df_fuel_consumption['MAKE'] == 'AUDI'] \
    .sort_values(by='FUELCONSUMPTION_HWY', ascending=True)
```

```python
# TOYOTA markasının COROLLA, AUDI markasının R8 modelini mean ve medianlarnı listeleyelim

(df_fuel_consumption
 .query('(MAKE == "TOYOTA" and MODEL.str.contains("COROLLA", case=False)) or (MAKE == "AUDI" and MODEL == "R8")')
 .groupby(['MAKE', 'MODEL'])[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG', 'CO2EMISSIONS']]                                                        
 .agg(['mean', 'median'])
)
```

```python
# Araçların MID tipinin ENGINESIZE, FUELCONSUMPTION_CITY, FUELCONSUMPTION_HWY, FUELCONSUMPTION_COMB, FUELCONSUMPTION_COMB_MPG özelliklerinin min ve max değerlerini listeleyelim

(df_fuel_consumption.query('VEHICLECLASS.str.contains("MID")')
    .groupby(['MAKE','MODEL','VEHICLECLASS'])[['ENGINESIZE', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']]
    .agg(['min','max']))
```

```python
df = pd.read_csv(filepath_or_buffer='data/drug200.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Yaş 40 den büyük olanların Age, Sex, BP, Cholesterol, Drug, 'Na_to_K Path I yöntemi ile getirelim
(df.query('Age > 40')
    .loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Drug', 'Na_to_K']]
    .sort_values(by='Age', ascending=False)
)
```

```python
# Yaş 20 den büyük olanların ve Sex M olanların Age, Sex, BP, Cholesterol, Drug, 'Na_to_K Path I yöntemi ile getirelim
(df.query('Age > 20 and Sex == "M"')
    .loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Drug', 'Na_to_K']]
    .sort_values(by='Age', ascending=False)
)
```

```python
# Sex = M olan yaşı 30 üzerinde olup 60 yaşının altında BP'si LOW olan Path I yöntemi ve iloc ile Age, Sex, BP, Drug getirelim
(df.query('Age > 30 and Age < 60 and Sex == "M" and BP == "LOW"')
    .iloc[:, [0,1,2,5]]
    .sort_values(by='Age', ascending=False)
)
```

```python
# Sex = F olan, yaşları 40 ve 40 dan büyük olan küçükten büyüğe sıralayarak, drugX ilacı kullanıp BP NORMAL ve Cholesterol HIGH olanların Age, Sex, BP, Cholesterol, Drug listeleyelim
(df.query('Sex == "F" and Age >=40 and Drug == "drugX" and Cholesterol == "HIGH" and BP == "NORMAL"')
    .loc[:, ['Age', 'Sex', 'BP', 'Cholesterol', 'Drug']]
    .sort_values(by='Age', ascending=True)
)
```
